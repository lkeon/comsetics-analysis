```python
import pandas as pd
```

```python
# Convert to dataframe
fileIn = 'face_oils_ingredients.json'
df = pd.read_json(fileIn).transpose().reset_index()
df.head()
```

```python
df.sort_values(by=['bestselling'], ascending=False, inplace=True)
```

```python
# Price of 50 bestselling oils in USD
ax = df[:50].plot.bar(x='index', y='price', figsize=(20,10))
ax.grid(axis='y')
ax.set_ylim(0, 100)
ax.set_xlabel('Face Oils with Descending Sales')
ax.set_ylabel('Price in USD');
```
